```python
import time
import random
import matplotlib.pyplot as plt
import numpy as np

# --- SIMULATION CONFIGURATION ---
TRUST_THRESHOLD = 0.85 # 85% of transactions come from trusted e-commerce nodes
BASE_HASH_TIME = 0.015 # Base time in ms for a basic cryptographic operation

print("==========================================================")
print(" E-COMMERCE BLOCKCHAIN PRE-DEVELOPMENT SIMULATION SUITE")
print("==========================================================\n")

# =====================================================================
# EXPERIMENT A: Transaction Scaling (Generates Data for Table 1)
# Fixed Network Size (N=100 nodes), Varying Transaction Volumes
# =====================================================================
print("--- RUNNING EXPERIMENT A: DATA FOR TABLE 1 ---")
tx_volumes = [100, 1000, 5000, 10000]
fixed_nodes = 100

for tx_count in tx_volumes:
    time_pbft, time_poa, time_trusted, time_unverified, time_blended = 0, 0, 0, 0, 0

    for _ in range(tx_count):
        node_trust = random.uniform(0, 1)

        # 1. PBFT O(N^2)
        time_pbft += (fixed_nodes * fixed_nodes) * (BASE_HASH_TIME / 150)

        # 2. Standard PoA O(N)
        time_poa += fixed_nodes * BASE_HASH_TIME

        # 3. Proposed Framework Bifurcation
        if node_trust >= TRUST_THRESHOLD:
            # Trusted Path O(1)
            time_trusted += BASE_HASH_TIME * 1.5
            time_blended += BASE_HASH_TIME * 1.5
        else:
            # Unverified Path O(N)
            time_unverified += fixed_nodes * BASE_HASH_TIME
            time_blended += fixed_nodes * BASE_HASH_TIME

    print(f"\nTransaction Volume: {tx_count} Tx")
    print(f"PBFT (O(N^2)):         {time_pbft:.1f} ms")
    print(f"PoA Baseline (O(N)):   {time_poa:.1f} ms")
    print(f"Proposed (Trusted):    {time_trusted if time_trusted > 0 else (BASE_HASH_TIME * 1.5 * tx_count):.1f} ms")
    print(f"Proposed (Unverified): {time_unverified if time_unverified > 0 else (fixed_nodes * BASE_HASH_TIME * tx_count):.1f} ms")
    print(f"Proposed (Blended):    {time_blended:.1f} ms")

print("\n" + "="*58 + "\n")

# =====================================================================
# EXPERIMENT B: Node Scaling (Generates Data & Plot for Figure 1)
# Fixed Tx Volume (1000 Tx), Varying Network Node Sizes
# =====================================================================
print("--- RUNNING EXPERIMENT B: GENERATING FIGURE 1 ---")
node_counts = [10, 50, 100, 150, 200]
fixed_tx = 1000

results_pbft, results_poa, results_blended, results_trusted = [], [], [], []

for N in node_counts:
    t_pbft, t_poa, t_blended, t_trusted = 0, 0, 0, 0
    for _ in range(fixed_tx):
        node_trust = random.uniform(0, 1)
        t_pbft += (N * N) * (BASE_HASH_TIME / 150)
        t_poa += N * BASE_HASH_TIME

        if node_trust >= TRUST_THRESHOLD:
            t_blended += BASE_HASH_TIME * 1.5
            t_trusted += BASE_HASH_TIME * 1.5
        else:
            t_blended += N * BASE_HASH_TIME
            t_trusted += BASE_HASH_TIME * 1.5 # Theoretical 100% trusted baseline

    results_pbft.append(t_pbft)
    results_poa.append(t_poa)
    results_blended.append(t_blended)
    results_trusted.append(t_trusted)

# PLOT GENERATION
plt.figure(figsize=(10, 6), dpi=300)
plt.plot(node_counts, results_pbft, marker='o', linestyle='-', color='#d62728', label='Traditional PBFT $\mathcal{O}(N^2)$')
plt.plot(node_counts, results_poa, marker='s', linestyle='--', color='#ff7f0e', label='Standard PoA $\mathcal{O}(N)$')
plt.plot(node_counts, results_blended, marker='^', linestyle='-', color='#2ca02c', linewidth=2, label='Proposed Architecture (Blended)')
plt.plot(node_counts, results_trusted, marker='D', linestyle=':', color='#1f77b4', label='Proposed Architecture (100% Trusted $\mathcal{O}(1)$)')

plt.yscale('log')
plt.title('Simulated Transaction Latency vs. Network Node Load (1,000 Tx Batch)', fontsize=14, pad=15)
plt.xlabel('Active Network Nodes (N)', fontsize=12)
plt.ylabel('Mean Latency in ms (Logarithmic Scale)', fontsize=12)
plt.legend(loc='upper left', frameon=True, shadow=True)
plt.grid(True, which="both", ls="--", alpha=0.4)
plt.tight_layout()
plt.show()
```
